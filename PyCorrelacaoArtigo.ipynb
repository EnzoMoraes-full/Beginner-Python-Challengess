```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt

# Conecta aos arquivos do drive
drive.mount('/content/drive')

# Lê o arquivo CSV
df = pd.read_csv('/content/drive/MyDrive/CorrelacaoArt/CorrelacaoArtigo.csv')

# Verifica Dataset
#print(df.head())

# Remove espaços e substitui vírgulas por pontos
df['Investimentos em pesquisa e desenvolvimento'] = df['Investimentos em pesquisa e desenvolvimento'].str.replace(',', '.').astype(float)
df['Quantidade de artigos'] = df['Quantidade de artigos'].str.replace(',', '.').astype(float)

print("")

# Calcula a correlação de Pearson
correlacao = df['Investimentos em pesquisa e desenvolvimento'].corr(df['Quantidade de artigos'])
print(f'Coeficiente de correlação: {correlacao}')

print("")

# Gráfico
plt.scatter(df['Investimentos em pesquisa e desenvolvimento'], df['Quantidade de artigos'])
plt.title('Relação entre Investimentos em P&D e Produção Científica')
plt.xlabel('Investimentos em P&D')
plt.ylabel('Produção Científica')
plt.grid()
plt.show()

```
